# Decision tree classifier ML

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
sb.set()
```

```python
data = pd.read_csv('labeled_data.csv')
data.head()
data.info()
```

```python
train_d, test_d = train_test_split(data, test_size = 0.20)
```

```python
Nervousness = pd.array(train_d['Nervousness_Label']).reshape(-1,1)
UTC = pd.array(train_d['Unable to control_Label']).reshape(-1,1)
Worry = pd.array(train_d['Worry_Label']).reshape(-1,1)
Trouble = pd.array(train_d['Trouble in Relaxation_Label']).reshape(-1,1)
Restlessness = pd.array(train_d['Restlessness_Label']).reshape(-1,1)
Irritability = pd.array(train_d['Irritability_Label']).reshape(-1,1)
Fear = pd.array(train_d['Fear_Label']).reshape(-1,1)



Stress1 = pd.array(train_d['Stress_Label']).reshape(-1,1)
```

```python
from sklearn.tree import DecisionTreeClassifier
dec_tree1 = DecisionTreeClassifier(max_depth=2)
```

# D tree for Nervousness

```python
dec_tree1.fit(Nervousness, Stress1)
```

```python
from sklearn.tree import export_graphviz
Nervousness_tree = export_graphviz(dec_tree1,
                         feature_names=None,
                         out_file=None,
                         filled=True,
                         rounded=True,
                         special_characters=True)
import graphviz
graphviz.Source(Nervousness_tree)
```

```python
Nervousness_pred_stress = dec_tree1.predict(Nervousness)
print("Classification Accuracy: ", dec_tree1.score(Nervousness, Stress1))

from sklearn.metrics import confusion_matrix
sb.heatmap(confusion_matrix(Stress1, Nervousness_pred_stress),
          annot=True,annot_kws={"size": 18})
```
