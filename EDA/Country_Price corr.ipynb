```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics
sb.set() # set the default Seaborn style for graphics
```

```python
data = pd.read_csv('winetrain.csv')
data.columns
```

```python
data.dtypes
```

```python
train, test = train_test_split(data, test_size=0.2)
train
```

```python
Price_Train = train.price.values.reshape(-1,1)
Price_Test = test.price.values.reshape(-1,1)
```

## boxplot for price

```python
sb.set(rc={'figure.figsize':(50,8)})
ax = sb.boxplot(x=train["price"])
```

## box plot of country with price

```python
sb.set(rc={'figure.figsize':(50,25)})
ax = sb.boxplot(x="country", y="price", data=train)
```
